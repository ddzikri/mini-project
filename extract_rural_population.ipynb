```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data/rural_population_percent.csv")
df
```

```python
# Menghitung jumlah data yang hilang sebelum dihapus
missing_data_count = df.isnull().sum()
print("Jumlah data yang hilang sebelum dihapus:")
print(missing_data_count)
print ("\n")
# Menghitung jumlah data duplikat sebelum dihapus
duplicate_data_count = df.duplicated().sum()
print("Jumlah data duplikat sebelum dihapus: ", duplicate_data_count)
```

```python
# Mengatasi data yang hilang
data_cleaned_rural = df.dropna()
print("Data yang hilang telah dihapus.")
print (data_cleaned_rural)
print ("\n")
# Menghapus data duplikat
data_cleaned_rural = df.drop_duplicates()
print("Data duplikat telah dihapus.")
print(data_cleaned_rural)
```

```python
data_cleaned_rural.to_csv("clean_data/cleaned_rural_population_percent.csv", index=False)
print("Data yang telah dibersihkan disimpan sebagai cleaned_rural_population_percent.csv")
```
