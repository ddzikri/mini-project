```python
import pandas as pd
```

```python
df = pd.read_csv("gdp_data.csv")
df
```

```python
# Menghitung jumlah data yang hilang sebelum dihapus
missing_data_count = df.isnull().sum()
print("Jumlah data yang hilang sebelum dihapus:", missing_data_count)
```

```python
# Menghitung jumlah data duplikat sebelum dihapus
duplicate_data_count = df.duplicated().sum()
print("Jumlah data duplikat sebelum dihapus:")
print(duplicate_data_count)
```

```python
# Mengatasi data yang hilang
data_gdp_missing = df.dropna()
print("Data yang hilang telah dihapus.")
print (data_gdp_missing)
```

```python
# Menghapus data duplikat
data_gdp_duplikat = df.drop_duplicates()
print("Data duplikat telah dihapus.")
print(data_gdp_duplikat)
```
