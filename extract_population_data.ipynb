## 2. EXTRACT POPULATION XML

```python
with open('raw_data/population_data.xml', 'r') as file:
    xml_data = file.read()
```

```python
# Membuat objek BeautifulSoup
soup = BeautifulSoup(xml_data, 'html.parser')
```

```python
data = []
for record in soup.find_all('record'):
    country = record.find('field', {'name': 'Country or Area'}).text
    item = record.find('field', {'name': 'Item'}).text
    year = record.find('field', {'name': 'Year'}).text
    value = record.find('field', {'name': 'Value'}).text
    
    # Cek jika year dan value tidak kosong sebelum mengonversinya menjadi integer
    if year.strip() and value.strip():
        data.append([country, item, int(year), int(value)])

xml = pd.DataFrame(data, columns=['Country', 'Item', 'Year', 'Value'])
```

```python
# Sesuaikan tipe data
xml['Year'] = pd.to_numeric(xml['Year'])
xml['Value'] = pd.to_numeric(xml['Value'])
```

```python
# Cek nilai yang hilang
missing_values = xml.isnull().sum()
print("Jumlah nilai yang hilang:")
print(missing_values)
print("\n")
# Cek duplikasi data
duplicate_data = xml.duplicated().sum()
print("Jumlah data duplikat:", duplicate_data)
```

```python
# Mengatasi data yang hilang
data_population_xml = xml.dropna()
print("Data yang hilang telah dihapus.")
print (data_population_xml)

# Menghapus data duplikat
data_population_xml = xml.drop_duplicates()
print("Data duplikat telah dihapus.")
print(data_population_xml)
```
